2.1. Grayscale với thư viện pillow

```python
from PIL import Image
from PIL import ImageShow
import matplotlib.pyplot as plt
import numpy as np


def to_grayscale_built(image):
  gray_image = []
  # Access pixels using `getdata()`
  pixels = image.getdata()
  # Process each pixel and build the grayscale image
  for pixel in pixels:
    gray_value = pixel[0] * 0.2126 + pixel[1] * 0.7152 + pixel[2] * 0.0722
    # Append to a new list (not gray_values directly)
    gray_image.append(gray_value)
    # Create a new grayscale image based on dimensions
  height = image.height
  width = image.width
  # Reshape the 1D list into a 2D array using NumPy
  gray_image_2d = np.array(gray_image).reshape(height, width)
  # Create a new grayscale image
  gray_image_2d_img = Image.fromarray(gray_image_2d.astype('uint8'))
  return gray_image_2d_img

# Open the image using Pillow
img = Image.open(io.BytesIO(response.content))
# Convert to NumPy array
img_np = np.array(img)
# Display the image using matplotlib
plt.imshow(img_np)
plt.axis("off")
# Crucial step: Show the plotted image
plt.show()

gray_image = to_grayscale_built(img)
gray_array = np.array(gray_image)
Image.fromarray(gray_array).save("grayscale_image.png")
# Hiển thị ảnh xám
# In ảnh xám
plt.imshow(gray_array, cmap='gray')
plt.axis("off")
plt.show()
```
